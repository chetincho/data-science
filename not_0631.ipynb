##### Vemos las distintas opciones que tenemos con la función tail()
- variable.tail(): muestra por defecto las últimas 5 filas de nuestro dataframe
- variable.tail(2): muestra las últimas dos filas de nuestro dataframe
- variable.tail(n): muestra las "n" últimas filas de nuestro dataframe

```python
df.tail()
```

```python
df.tail(2)
```

```python
df.tail(50)
```
